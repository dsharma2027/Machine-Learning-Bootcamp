```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

Functional Method of Plotting

```python
plt.plot(x,y)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Title")
plt.show()   # not required in Jupyter notebook but required in any other IDE
```

```python
# Adding subplot

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
plt.show()
```

Object oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.7])   #([left, bottom, width, height])  
axes.plot(x,y)
axes.set_xlabel("X-Label")
axes.set_ylabel("Y-Label")
axes.set_title("Title")
plt.show()
```

```python
# Subploting through Objet oriented  plotting

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")
axes2.plot(y,x)
axes2.set_title("Smaller Plot")
plt.show()
```

```python
# Using Subplot object instead of figue for creating a sub plot

fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")
plt.tight_layout()   # will remove in overlapping
plt.show()

# for curr_ax in axes:
    #curr_ax.plot(x,y)
```

# Figure Size,Aspect Ratio and DPI

```python
fig = plt.figure(figsize=(3,2))   # DPI we are taking as by default

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
plt.show()
```

Legends

```python
fig1 = plt.figure(figsize=(6,4),dpi=100)

ax = fig1.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Square")
ax.plot(x,x**0.5,label="X Root")
ax.legend(loc=0)
plt.show()
```

# Plot Appearance

```python
# Color and styling the plpt of Plot Line

fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=5,alpha=0.6,linestyle='-',marker='o')   
# you can also add RGB Hex code for custom color 
# Linestyle is used for type of line 
# alpha parameter decide the transparency of the line
# Markers are used for marking the point occured 
plt.show()
```

# Plot Limits

```python
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])
ax.plot(x,y, lw=4,ls='--')

ax.set_xlim([0,1])  # Setting the limit on x axis
ax.set_ylim([0,2])

plt.show()
```
